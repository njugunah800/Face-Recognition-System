```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload an image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read the image
image = cv2.imread(image_path)

if image is None:
    raise ValueError("Image could not be read. Check the file format or try re-uploading.")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load Haar Cascade Classifier
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,       # Balanced detection
    minNeighbors=6,        # Higher value = more confidence
    minSize=(50, 50)       # Avoid detecting small false positives
)

# Filter faces by aspect ratio (optional, to exclude false rectangles like shoulders)
filtered_faces = []
for (x, y, w, h) in faces:
    aspect_ratio = w / float(h)
    if 0.75 < aspect_ratio < 1.3:  # Typical face aspect ratio range
        filtered_faces.append((x, y, w, h))

print(f"Filtered and confirmed faces: {len(filtered_faces)}")

# Draw rectangles and number them
for i, (x, y, w, h) in enumerate(filtered_faces):
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(image, f'Face {i+1}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

# Convert BGR to RGB for matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Show image with bounding boxes
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Detected Faces")
plt.show()

# Save result
cv2.imwrite("detected_faces.jpg", image)
```
